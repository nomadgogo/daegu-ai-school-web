```python
plt.subplot(2,1,1)      #두개 중 첫번째 꺼
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)        #두개 중 두번째 꺼
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')   #짧은 색상 코드 (g=green, 블랙=k)
plt.plot(x, np.sin(x-2), color='0.75')   # 1=흰색, 0=진한색   (0~1 조절)
plt.plot(x, np.sin(x-3), color='#ffdd44') 
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))   #십진수로.. RGB 빛의 3원색....
plt.plot(x, np.sin(x-5), color='chartreuse') #색의 이름으로...
```

```python
plt.plot(x, x+0, linestyle='solid')   #실선
plt.plot(x, x+1, linestyle='dashed')  # 점섬
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')   
plt.plot(x, x+5, linestyle='--')  
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0,'-g')   #조합 설정 가능
plt.plot(x, x+1,'--c')
plt.plot(x, x+2,'-.k')
plt.plot(x, x+3,':r')
```

```python
plt.plot(x, np.sin(x),'-g', label='sin(x)')
plt.plot(x, np.cos(x),':b', label='cos(x)')
plt.axis('equal')

plt.legend()   # 주석
```

```python
x = np.linspace(0,10,30)
plt.plot(x, np.sin(x), 'o', color='black')   #산점도
```

```python
rng = np.random.RandomState(0)
for marker in ['o',',',',','x','+','v','^','<','>','s','d']:  #산점도의 점들을 marker라고 하는데, 그 마커들을 여러가지로 표현가능
    plt.plot(rng.rand(5),rng.rand(5), marker, 
             label='marker={0}'.format(marker))  # 각 마커를 서령하는 주석표시
    plt.legend()
```

```python
y = np.sin(x)
plt.plot(x, y, '-ok')  #-선 연결, k블랙
```

```python
plt.plot(x,y, '-p', color='gray',
        markersize=15, linewidth=4,
        markerfacecolor='white',
        markeredgecolor='red',
        markeredgewidth='2')      # 마커 사이즈, 선 굵기도 조절, 마커 속을 흰색으로, 마커 모퉁이의 색&굵기 조절
```

```python
plt.scatter(x,y)   # scatter도 똑같이 원으로 표현,데이터 분포를 보여줄 때는 활용도가 더 높음
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)  #x에 랜덤으로 100개의 값을 찍힘 (랜덤 -3 ~ 3 사이)
y = rng.randn(100)  #y에 랜덤으로 100개의 값을 찍힘
color = rng.rand(100)   # 컬러도 랜덤으로 다양하게...
sizes = 700 * rng.rand(100)  # 700 곱해서 크기 다양하게 표현...

plt.scatter(x, y, c=color, s=sizes, alpha=0.3, cmap='viridis')    # alpha 투명도 조절
plt.colorbar()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
```

```python
# features      #싸이키런의 배열도 넘파이 배열로 저장되어 있음을 확인
# features.shape  #4 개의 속성, 150개의.. 을 확인 
plt.scatter(features[0], features[1], alpha=0.2, 
           s=features[3]*100, cmap='viridis',      #피쳐3을 기준으로 사이즈를 조절
           c=iris.target)                            # 타겟(이아리스 꽃 종류)에 따라서 구분


plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
#errorbar
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')   # y축으로 에러의 범위를 표현 dy=0.8
```

```python
plt.style.use('seaborn-whitegrid')                           #스타일적용 
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
            ecolor='lightgray', elinewidth=3)
```
